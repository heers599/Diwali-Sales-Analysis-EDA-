# EXPLORATORY DATA ANALYSIS

### Gender

```python
di.columns
```

```python
sns.countplot(x = 'Gender',data = di)
```

```python
f_m = sns.countplot(x = 'Gender',data = di)
for bars in f_m.containers:
    f_m.bar_label(bars)
```

```python
di.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen=di.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

#### Age

```python
f_m = sns.countplot(x = 'Age Group',data = di , hue = 'Gender')
for bars in f_m.containers:
    f_m.bar_label(bars)
```

```python
di.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_age=di.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

### State

```python
sales_state=di.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x='State', y='Orders', data = sales_state)
```

```python
sales_state=di.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State', y='Amount', data = sales_state)
```

### Marital- Status

```python
f_m = sns.countplot(x = 'Marital_Status',data = di)
sns.set(rc={'figure.figsize':(2,5)})
for bars in f_m.containers:
    f_m.bar_label(bars)
```

```python
sales_state=di.groupby(['Marital_Status', 'Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(2,5)})
sns.barplot(x='Marital_Status', y='Amount', data = sales_state, hue = 'Gender')
```

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
f_m = sns.countplot(data = di, x = 'Occupation')

for bars in f_m.containers:
    f_m.bar_label(bars)
```

```python
sales_state=di.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data = sales_state)
```

### Product category

```python
sns.set(rc={'figure.figsize':(30,5)})
f_m = sns.countplot(data = di, x = 'Product_Category')

for bars in f_m.containers:
    f_m.bar_label(bars)
```

```python
sales_state=di.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='Product_Category', y='Amount', data = sales_state)
```

```python
sales_state = di.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
di.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
